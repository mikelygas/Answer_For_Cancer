```python
#Cleaning CSV Files

breast_mortality_global_crude = breast_mortality_global_crude.dropna()
breast_mortality_global_crude.head()

breast_new_cases_global_crude = breast_new_cases_global_crude.dropna()
breast_new_cases_global_crude.head()
```
